```
import pandas as pd
```

```
df = pd.read_csv('/content/AB_NYC_2019.csv')
```

# Deleting The column with with missing data

```
df1 = df.copy()
```

```
df1.head()
```

```
df1.isnull().sum()
```

```
df1.drop("last_review",axis =1,inplace=True)
```

```
df1.isnull().sum()
```

# 2 DELETING ROW WITH MISSING DATA

```
df2 = df.copy()
```

```
df2.dropna(inplace=True)
```

```
df2.isnull().sum()
```

```
df2.shape
```

# 3  Filling values with imputation

```
df3 = df.copy()
```

```
df3.head()
```

# Filling Categorical Data
#Use mode value

```
df3.nunique()
```

```
df3["last_review"].fillna(df3["last_review"].value_counts().index[0],inplace=True)
```

```
df3.isnull().sum()
```

# Categorical data Using it as new category

```
df4=df.copy()
```

```
df4.isnull().sum()
```

```
df4["last_review"].fillna("Not_Reviewed",inplace=True)
```

```
df4.sample(10)
```

# Advanced imputation

```
df5=df.copy()
```

```
df5.isnull().sum()
```

```
df5["reviews_per_month"].interpolate(inplace=True)
```
